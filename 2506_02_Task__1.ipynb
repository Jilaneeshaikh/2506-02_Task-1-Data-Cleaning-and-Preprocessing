**Handling Missing Values**

```python
df.dropna(subset=["date_added"], axis=0, inplace=True)
df.shape
```

```python
round(df.isnull().sum()/df.shape[0]*100,2).sort_values(ascending=False)
```

```python
#dropping rows for small percentage of null
df.dropna(subset=["rating","duration","date_added"], axis=0, inplace=True)
df.shape
```

```python
round(df.isnull().sum()/df.shape[0]*100,2).sort_values(ascending=False)
```

# **Duplicate Value**

```python
# Step 5: Remove Duplicates
print("\nDuplicate Rows:", df.duplicated().sum())
df = df.drop_duplicates()
```


## **Replace Missing Value**

```python
df["country"].replace(np.nan, "Unknwon", inplace=True)
df["cast"].replace(np.nan, "No Cast", inplace=True)
df["director"].replace(np.nan, "No Director", inplace=True)
round(df.isnull().sum()/df.shape[0]*100,2).sort_values(ascending=False)
```

```python
movies_df=df.loc[df["type"]=="Movie"]

movies_df.head(2)
```

```python
shows_df=df.loc[df["type"]=="TV Show"]
shows_df.head(2)
```

```python
# Optional: Clean release_year
df['release_year'] = df['release_year'].astype(int)
```

```python
# Step 10: Export Cleaned Data
df.to_excel("Netflix_cleaned.xlsx", index=False)
print("\nCleaned dataset saved as 'Netflix_cleaned.xlsx'")
```
